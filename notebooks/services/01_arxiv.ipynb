# arXiv client — service test

Tests `ArxivClient`: fetch metadata + download a PDF.

**Prereqs:** none (pure external API). 
**⚠️ Rate limit:** arXiv allows ~1 req / 3s. Don't re-run cells rapidly — cumulative requests get you banned for minutes.

```python
from dotenv import load_dotenv
load_dotenv()

from src.services.arxiv.factory import make_arxiv_client

client = make_arxiv_client()
client
```

```python
# Fetch a few recent cs.AI papers (uses top-level await — Jupyter supports it)
papers = await client.fetch_papers(max_results=3)

for p in papers:
    print(p.arxiv_id, '|', p.title[:70])
print(f"\n{len(papers)} papers fetched")
```

```python
# Inspect one paper's full metadata shape
papers[0].model_dump()
```

```python
# Download the PDF (cached under ./data/arxiv_pdfs — second run is instant)
pdf_path = await client.download_pdf(papers[0])
print('saved to:', pdf_path)
print('size KB:', pdf_path.stat().st_size // 1024)
```
