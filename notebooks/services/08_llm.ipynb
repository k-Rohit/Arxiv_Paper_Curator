# OpenAI LLM (chat-completion) — service test

Tests `OpenAIClient.generate_rag_response`: prompt building + grounded answer.

**Prereqs:** `OPENAI_API_KEY` in `.env`. Costs ~$0.005 per run.

```python
from dotenv import load_dotenv
load_dotenv()

from src.services.openai_.factory import make_openai_client

llm = make_openai_client()
llm
```

```python
# Fake retrieved chunks (no OpenSearch needed — LLM test in isolation)
chunks = [
    {
        'arxiv_id': '1706.03762',
        'chunk_text': ('Multi-head attention allows the model to jointly attend to '
                       'information from different representation subspaces at different positions. '
                       'We employ h=8 parallel attention heads.'),
    },
    {
        'arxiv_id': '1706.03762',
        'chunk_text': ('Scaled dot-product attention computes the dot products of the query '
                       'with all keys, divides each by sqrt(dk), and applies softmax.'),
    },
]

result = await llm.generate_rag_response(query='How does multi-head attention work?', chunks=chunks)
print(result.get('answer', result))
```

```python
# Groundedness check: ask something the chunks DON'T cover —
# a good RAG prompt should make the LLM admit it doesn't know
result = await llm.generate_rag_response(query='What optimizer did they use for training?', chunks=chunks)
print(result.get('answer', result))
```
